# Reporte sobre GIT
___

**GIT** es un sistema de control de versiones, por lo tanto es un software que permite a los desarrolladores mantener un historial completo de cada archivo en un proyecto. Sin embargo, no se necesita ser un desarrollador para usar **GIT**; por ejemplo, puedes usarlo para mantener un historial de todos los archivos de un sitio y cada vez que se cambias uno, se registra una versión del cambio y de ser necesario siempre puedes volver a cualquier versión y restaurarla.

Puede llegarse a confundir que **GIT** se utilice como simplemente una herramienta de copia de seguridad, pero hace mucho más que eso, no solo almacena un historial completo de tus archivos, sino que también almacena la 'versión' de esos archivos a lo largo del historial de un proyecto. Esto te permite usarlo para recuperar información sobre cuándo se editó un archivo, así como la capacidad de ver los cambios exactos realizados en el archivo. Una copia de seguridad simple no incluye estos detalles adicionales.

Nos ayudar a obtener la siguiente informacion:

1. ¿Quién escribió este módulo? 
2. ¿Cuándo se editó esta línea en particular de este archivo en particular? ¿Por quién? ¿Por qué fue editado?
3. Durante las últimas 1000 revisiones, ¿cuándo / por qué dejó de funcionar una prueba ude regresion?

## Comandos mas usados
___
### Comandos Basicos
| Comando | Description |
| ------ | ----------- |
| git help \<command>  | ayuda del comando GIT a usar |
| git init | crea un nuevo repositorio  |
| git status    | Muestra el estatus actual |
| git add \<filename>    | agrega archivos al área de preparación |
| git commit    | crea un nuevo commit |
| git log    | muestra la historia en lista |
| git log --all --graph --decorate    | emuestra la historia simulando forma de arbol |
| git diff \<filename>    | mostrar los cambios que realizó en relación con el que esta en staging |
| git diff \<revision> \<filename>    | muestra diferencias en un archivo entre instantáneas|
| git checkout \<revision>    | Actualiza el HEAD y la rama actual|

    
### Branching y Merging  
| Comando | Description |
| ------ | ----------- |
|git branch| muestra los branches |
|git branch \<name>| Crea un nuevo branch|
|git checkout -b \<name>| Crea un nuevo branch y lo cambia al actual|
|git merge \<revision>| Hace un Merge con el que cambiamos |
|git mergetool| Usa la herramienta default para tratar de hacer un Merge |
    
### Remoto  
| Comando | Description |
| ------ | ----------- |
|git remote| lista los remotos |
|git remote add \<name> \<url>| agrega un remoto |
|git push \<remote> \<local branch> : \<remote branch> | enviar objetos a remoto y actualizar la referencia remota |
|git branch --set-upstream-to=<remote>/<remote branch>| Establecer correspondencia entre la version local y remota |
|git fetch| recuperar objetos / referencias del remoto |
|git pull| lo mismo que git fetch; git merge |
|git clone| descargar repositorio desde un remoto |

### Deshacer
| Comando | Description |
| ------ | ----------- |
|git commit --amend| Edita el comentario que se hara en el commit |
|git reset HEAD \<file>| desmontar un archivo |
|git checkout -- \<file>| descartar los cambios |
